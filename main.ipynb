아래 x1,y1,x2,y2가 회절선의 밝은 부분 한 절단의 양 끝의 좌표이고 두 점 사이 거리 공식을 이용해 사진 내에서의 길이(픽셀)를 구해주었다.
<br> 그 후 사진 내의 자와 비례식을 이용하여 cm로 길이를 구해주었다. <br>그 후 공식을 이용하여 용수철의 두께를 구했다. 

```python
x1,y1,x2,y2 = (988,550,980,546)
X = pow((x1-x2)**2 + (y1-y2)**2,1/2)
d = X/52.1
D = 143
lamda = 6.5 * pow(10,-5)

용수철두께 = lamda*D/d * 10
print('용수철의 두께 : ' + str(용수철두께)[:6] + 'mm')
```
